```r
x= NULL
x[1]= 0

for(i in 2:1000){
x[i] = x[i-1] + rnorm(1)
}
print(x)
```

```r
random_walk = ts(x)

par(bg='white')

plot(random_walk, main = 'A Random Walk', ylab = '', xlab = 'Days', col= 'red', lwd = 2)
```

```r
par(bg='white')
acf(random_walk)



# Graph shows High Correlation (i.e. Non Stationary)
```

```r
diff(random_walk)
```
